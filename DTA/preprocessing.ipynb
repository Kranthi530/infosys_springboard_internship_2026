# NLP Text Preprocessing on Unstructured Web Data

This notebook demonstrates how to clean and preprocess messy HTML text collected from web scraping.

The raw data contains:
- HTML tags
- JavaScript code
- URLs
- Emojis
- Special characters
- Duplicate sentences
- Noise like hashtags and usernames

The preprocessing pipeline includes:

1. Removing HTML tags
2. Removing JavaScript code
3. Removing URLs
4. Lowercasing text
5. Removing special characters
6. Tokenization
7. Stopword removal
8. Lemmatization

The final result is clean text ready for NLP tasks such as sentiment analysis and topic modeling.

## Step 1: Import Libraries

We import necessary Python libraries for text preprocessing.

- BeautifulSoup → remove HTML tags
- re → regular expressions
- nltk → tokenization, stopwords, lemmatization

```python
import re
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

## Step 2: Download NLP Resources
NLTK requires additional datasets for tokenization and stopword removal.

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

## Step 3: Raw HTML Input

The dataset contains raw HTML scraped from web pages.
It includes tags, scripts, emojis, URLs, and noisy symbols.

```python
raw_text = """
<p>
Just got the new Apple Vision Pro, and it's amazing!
The spatial computing is mind-blowing.
But there are data privacy concerns.
</p>

<p>
Customer reviews are mixed.
One user wrote "The AR features were a game changer!!!"
Another said "It is a waste of money $$$."
</p>

<a href="https://www.theverge.com">https://www.theverge.com</a>
"""
```

## Step 4: Remove HTML Tags

HTML tags like `<p>`, `<div>`, `<script>` are not useful for NLP.
We remove them using BeautifulSoup.

```python
soup = BeautifulSoup(raw_text, "html.parser")
clean_text = soup.get_text()

print(clean_text[:500])
```

## Step 5: Remove URLs

Web scraped text often contains URLs that are not useful for analysis.
We remove them using regular expressions.

```python
clean_text = re.sub(r"http\S+|www\S+", "", clean_text)
```

## Step 6: Remove Code Snippets

Scraped pages may include JavaScript fragments such as:
console.log(), var pageTitle etc.

We remove such patterns using regular expressions.

```python
clean_text = re.sub(r"console\.log.*", "", clean_text)
clean_text = re.sub(r"var\s+\w+.*", "", clean_text)
```

## Step 7: Convert Text to Lowercase

Lowercasing ensures that words like "Apple" and "apple" are treated as the same word.

```python
clean_text = clean_text.lower()
```

## Step 8: Remove Special Characters

The text includes symbols such as:
- emojis
- dollar signs
- encoding artifacts like Â, Ã

These are removed using regular expressions.

```python
clean_text = re.sub(r"[^a-zA-Z\s]", " ", clean_text)
```

## Step 9: Remove Extra Spaces
After cleaning symbols, multiple spaces may appear.
We normalize them.

```python
clean_text = re.sub(r"\s+", " ", clean_text).strip()

print(clean_text[:200])
```

## Step 10: Tokenization

Tokenization splits text into individual words.

```python
tokens = word_tokenize(clean_text)

print(tokens[:30])
```

## Step 11: Stopword Removal

Stopwords are common words like:
- the
- is
- in
- and

They do not contribute much meaning for NLP tasks.

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens[:30])
```

## Step 12: Lemmatization

Lemmatization converts words into their base form.

Examples:
running → run  
studies → study  
better → good

```python
lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print(lemmatized_tokens[:30])
```

## Step 13: Final Clean Text

After preprocessing, we join the tokens back into a clean sentence.
This text is now ready for NLP tasks such as sentiment analysis, classification, or topic modeling.

```python
final_text = " ".join(lemmatized_tokens)

print(final_text[:500])
```

```python
print("Total tokens after cleaning:", len(lemmatized_tokens))
```
